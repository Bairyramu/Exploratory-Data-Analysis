# Gaussion Distribution

```python
figure=data.Age.hist(bins=50)
figure.set_title("Age")
figure.set_xlabel("Age")
figure.set_ylabel("No of passenger")
```

```python
figure=data.boxplot(column="Age")
```

```python
data["Age"].describe()
```

```python
## Assuming Age follows A gaussion Distribution we will calculate the boundries wchich differentiates the outliers
upper_boundary=data["Age"].mean()+3*data["Age"].std()
lower_boundary=data["Age"].mean()-3*data["Age"].std()
```

```python
print("Upper boundary: ",upper_boundary)
print("Lower boundary: ",lower_boundary)
print("Mean: ",data["Age"].mean())
```

```python
# IQR 
IQR=data.Age.quantile(0.75)-data.Age.quantile(0.25)
```

```python
IQR
```

```python
lower_bridge=data["Age"].quantile(0.25)-(IQR*1.5)
```

```python
lower_bridge
```

```python
upper_bridge=data["Age"].quantile(0.75)+(IQR*1.5)
```

```python
upper_bridge
```
